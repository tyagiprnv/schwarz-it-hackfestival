```python
import pandas as pd
from datetime import datetime, timedelta
import json

# Function to determine if an item is perishable based on packaging
def is_perishable(packaging):
    perishable_packaging = ['bag', 'punnet', 'each', 'bunch', 'carton', 'block', 'tub', 'cup']
    return packaging in perishable_packaging

# Load the inventory dataset
file_path = '/Users/ayushmukherjee/Desktop/HackFestivalCode/hackathon_carbon_dataset.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Convert to DataFrame
inventory_df = pd.DataFrame(data)

# Convert 'expiresAt' to datetime format
inventory_df['expiresAt'] = pd.to_datetime(inventory_df['expiresAt'])

# Get today's date
today = datetime.today()

# Add a column for the number of days to expiration
inventory_df['daysToExpiration'] = (inventory_df['expiresAt'] - today).dt.days

# Create empty DataFrames for the different categories
discounted_df = pd.DataFrame()
donation_df = pd.DataFrame()

# Iterate through inventory
for index, row in inventory_df.iterrows():
    packaging = row['packagingUnit']
    days_left = row['daysToExpiration']
    
    if is_perishable(packaging):
        # Perishable Items - 5 days for discount, 2 days for donation
        if days_left == 5:
            discounted_df = pd.concat([discounted_df, pd.DataFrame([row])], ignore_index=True)
        elif days_left == 2:
            donation_df = pd.concat([donation_df, pd.DataFrame([row])], ignore_index=True)
    else:
        # Non-Perishable Items - 4 days for discount, 2 days for donation
        if days_left == 4:
            discounted_df = pd.concat([discounted_df, pd.DataFrame([row])], ignore_index=True)
        elif days_left == 2:
            donation_df = pd.concat([donation_df, pd.DataFrame([row])], ignore_index=True)

# Save discounted and donation lists to CSV files for further actions
discounted_df.to_csv('discounted_items.csv', index=False)
donation_df.to_csv('donation_items.csv', index=False)

# Printing summary of results
print("Items marked for discount:")
print(discounted_df)

print("\nItems marked for donation:")
print(donation_df)
```

```python
import pandas as pd
import json
import requests

# Load historical purchase data and inventory data
historical_data_path = '/Users/ayushmukherjee/Desktop/HackFestivalCode/historical_purchase_data.csv'
inventory_data_path = '/Users/ayushmukherjee/Desktop/HackFestivalCode/hackathon_carbon_dataset.json'

# Load the historical purchase data (CSV format)
historical_df = pd.read_csv(historical_data_path)

# Load inventory data (JSON format)
with open(inventory_data_path, 'r') as f:
    inventory_data = json.load(f)

# Prepare the inventory data for recommendation use
inventory_items = {item['name']: item for item in inventory_data}

# Function to get the most frequent purchases by the organization
def analyze_purchase_history(org_id, historical_df):
    # Filter purchases for the organization
    org_purchases = historical_df[historical_df['id'] == org_id]
    
    # Extract the 'purchase' column, which contains a list of items
    frequent_items = org_purchases['purchase'].apply(lambda x: eval(x))  # Parse the purchase JSON
    
    # Create a list of items that the organization buys frequently
    item_list = []
    for purchases in frequent_items:
        for purchase in purchases:
            item_list.append(purchase['item'])

    # Get the most frequent items
    frequent_item_count = pd.Series(item_list).value_counts().head(5)  # Top 5 most purchased items
    return frequent_item_count.index.tolist()

# Function to generate targeted advertisements for each organization using phy3 by Ollama
def generate_targeted_ads(org_id, org_name, frequent_items, inventory_items):
    inventory_string = ', '.join(inventory_items.keys())  # List all inventory items
    
    prompt = f"""
    You are an AI assistant that provides targeted advertisement recommendations for organizations based on their purchase history.
    The organization, {org_name}, has a history of purchasing the following items: {', '.join(frequent_items)}.
    
    Based on the available inventory which includes: {inventory_string}, suggest specific items that the organization would likely be interested in for their next purchases.
    Provide a brief reason why each item would be suitable based on their past purchases.
    """
    
    # Call phy3 model via Ollama API (make sure Ollama is set up and running)
    response = requests.post(
        'http://localhost:11434/api/generate', 
        headers={"Content-Type": "application/json"},
        data=json.dumps({
            "prompt": prompt,
            "model": "phi3",  # Specify the model here
            "stream": False
        })
    )
    
    if response.status_code == 200:
        response_json = response.json()  # Get the full response as a JSON object
        # Check if the 'completion' field exists
        if 'completion' in response_json:
            return response_json['completion']
        else:
            # If the 'completion' key doesn't exist, return the entire response for debugging
            return f"Unexpected response structure: {response_json}"
    else:
        return f"Error: {response.status_code} - {response.text}"

# Iterate through each organization in historical data and generate ads
def create_targeted_ads_for_all_organizations(historical_df, inventory_items):
    org_ads = {}
    org_ids = historical_df['id'].unique()

    for org_id in org_ids:
        org_name = historical_df[historical_df['id'] == org_id]['company_name'].iloc[0]
        frequent_items = analyze_purchase_history(org_id, historical_df)
        ad = generate_targeted_ads(org_id, org_name, frequent_items, inventory_items)
        org_ads[org_name] = ad
        break
    return org_ads

# Generate ads for all organizations
targeted_ads = create_targeted_ads_for_all_organizations(historical_df, inventory_items)


# Display the targeted ads for each organization
for org, ad in targeted_ads.items():
    print(f"Targeted Ads for {org}:")
    print(ad)
    print("\n" + "="*80 + "\n")
```
